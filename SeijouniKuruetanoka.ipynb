```python
import os, json, requests

songs_lyrics = ""
songs = requests.get("https://subekashi.izmn.net/api/song/?format=json").json()
for song in songs :
    if song["lyrics"] :
        if song["isjapanese"] and (not song["isjoke"]) and ("\r\n" in song["lyrics"]):
            song_lyrics = song["lyrics"]
            songs_lyrics += song_lyrics.replace("\r\n", "<|endoftext|>")
            songs_lyrics += "<|endoftext|>"
        else :
            print(song["title"], song["id"], ":", song["isjapanese"], not song["isjoke"], "\r\n" in song["lyrics"])

os.makedirs("subeana", exist_ok=True)
with open('subeana/output.txt', 'w') as f:
    json.dump(songs_lyrics, f, ensure_ascii=False)
```
